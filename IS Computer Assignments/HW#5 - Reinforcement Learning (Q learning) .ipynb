        The <b> concept of delayed gratification (reward)</b>  and the <b> crucial role </b> it plays in <b> decision making </b> in human being lives, makes it a topic that needs a brief investigation before its used in a "Reinforcement Learning" framework.<br><br> Situations where we have to choose between two options that  have <b> a trade-off between immediate gratification and long-term</b>  (but larger) gains are <b> very frequent in our lives</b> , which is why the decision-making processes to delay gratification have been studied extensively throughout different developmental ages, like the <b> famous marshmallow test indicating children's developmental age. </b>How we <b> deal with delayed rewards</b> , depends on various aspects: <b> environment dynamics, our myopic of hyperopic view, reward type, ...</b>  <br><br>
        
        
<p style = "font-family:georgia;">

<font color= #00000><font>
<p style = "font-family:cambria;">
    <font size="3">

        The problem with this strategy is that if<b>  two completely different paths,</b>  one leading to the goal and one leading to very negative rewards <b> both pass a certain state with a high instant reward </b> the <b> agent won't be prevented </b> to take the <b> bad path unless it experiences it until the end </b> for each single episode. In other words <b> the agent cant use it's previous experience to prevent repeating a mistake </b><br><br></b> <br>If the agent chooses actions, <b> with a farsighting view based on all the rewards its going to get from now on,</b> The <b> next rewards </b> the agent will collect, can be modeled in a diminishing ( discounted return ) or equal (average reward ) fashion </b><br>
    
<p style = "font-family:georgia;">
    <b><font size="4"> <font color="#D6274F" >  Discounted Return  </b><font></font>
    </font>    
</p>
    
<font color= #00000><font>
<p style = "font-family:cambria;">
    <font size="3">
<br> $$\boxed{\mathit{G_{t} = R_{t}\;+ \gamma R_{t+1}\;+ \gamma^{2} R_{t+2}\;+\;...} }$$ <br><br>
Considering Discounted Return <b> makes previous experiences helpful and also ensures the target function's convergence </b><br><br><br>

These methods are usually used when the environment isnt experienced before, thus $\mathit{P_{ss'}^{\,a}}$ and $\mathit{R_{ss'}^{\,a}}$ aren't used.<br><br><br>In <b> Contrast</b>  to model-based learning, <b>‌ model-free learning</b>, learns <b> value fucntions and then policies, or even learns policies directly ( Policy Gradient Methods )

```python
ERROR_RATE = 0.05
MIN_EPISODES = 3000
```

## Building Environment:

### Reading ENV.map, saving environment data:

```python
import numpy as np

def read_env(file_name):
    
    with open(file_name, 'r') as f:
        env = [[int(num) for num in line.split(',')] for line in f]
    env = np.array(env)
    return env

env = read_env('ENV.map')
nA = 2*len(env.shape) # action numbers = 2 (moving Forward and Backward) in each dimension

LENGTH = env.shape[1]
WIDTH = env.shape[0]

env = np.ndarray.flatten(np.array(env))
nS = len(env) # fetching state numbers 
```

```python
states = np.arange(nS*nS) + 1 # Generating states for the environment
```

## Each Environment has 2 elements representing MDP Dynamics:
1. $\mathit{\;\;R_{ss'}^{\,a}}$ <br><br>
2. $\mathit{\;\;P_{ss'}^{\,a}}$ 

### $\mathit{\;\;R_{ss'}^{\,a}}$ in our problem only depends on our new position in the grid and is only a function of "next state".<br><br>    In other words:

$\mathit{\;\;R_{ss'}^{\,a}\,=\,f\;(s')}$

### Assigninig Rewards to states with respect to their type:

```python
Rss = np.zeros(nS) #initializing Rewards:

start = np.where(env == 1) # Reward: 0(?)
# Good states:

T = np.where(env == 3) # Reward: 100 (Terminal State (Goal))
Rss[T] = 100
r2 = np.where(env == 4) # Reward: 0 or 30 (Rewards that can help moving towards the Goal)
Rss[r2] = 0

# Bad states:

c1 = np.where(env == 0)# Reward: -0.01 (Punishment; preventing too much exploration, short path is better)
Rss[c1] = -0.01

c2 = np.where(env == 2) # Reward: -20 (Punishment;  Soft obstacles)
Rss[c2] = -20

c3 = np.where(env == 5) # Reward: -100 (Punishment; Hard  obstacles)
Rss[c3] = -100

```

### $\mathit{\;\;P_{ss'}^{\,a}}$ is implemented with the given error rate = 5% between decisions and actions in the following function:

#### This function generates the next state for a current state by first mapping decisions (based on policy and optimal action) to actions, and then generating the next_state based on action direction:

```python
def next_state_gen(current_state, decision): 
    
    done = False #Termination Flag
    
    # DECISION: the OPTIMAL action referred in POLICY
    # ACTION: can be DIFFERENT because of ERROR
    
    # Mapping Decisions to Actions with error = ERROR_RATE:
    
    def decision_to_action(decision):
        
        mapping_pr = np.ones(nA)*ERROR_RATE/(nA-1)
        mapping_pr[decision] = 1-ERROR_RATE
        #print(mapping_pr)
        action = np.random.choice(nA, p = mapping_pr)
        return action
    
    action = decision_to_action(decision)
    
    action_to_move = {
    0: -LENGTH, #moving UP
    1: 1,  #moving RIGHT
    2: LENGTH, #moving DOWN
    3: -1  #moving LEFT
    }
    next_state = current_state + action_to_move[action]
    
    #considering horizontal walls:
    if((next_state < 1) or (next_state > nS)):
        #print("Vertical Block\n")
        next_state = current_state
   
    #considering vertical walls:
    if(((current_state % LENGTH == 0) and (action == 1)) or ((current_state % LENGTH == 1)) and (action == 3)):
        #print("Horizontal Block\n")
        next_state = current_state
    
    #informing us in case of reaching terminal state:
    if(next_state == T[0][0] + 1):
        done = True
    return next_state,done

```

### This function chooses an action according to the policy the agent is following,
### that itself is a function of q-values

```python
def make_decision(state_policy): #gets policy (action pr's) for current state, makes decision with sampling
    decision = np.random.choice(nA, p = state_policy)
    return decision
```

### Implementing Soft-max and E-greedy, 
### We feed them with q_values[state] when we call them, In other words, they process q_values locally !

```python
def soft_max(q_values, beta = 1): #Beta = 1/T (T is used in boltzam, they are the same policy)
    
    pr_actions = np.zeros(nA)
    print(f"Softmax Numerators: {np.exp(q_values*beta)}\n")
    pr_actions = ((np.exp(q_values*beta))/(np.sum(np.exp(q_values*beta))))
    return pr_actions


def epsilon_greedy(q_values,epsilon =0.1):
    
    optimal_action  = np.argmax(q_values)
    pr_actions = (epsilon* np.ones(nA))/nA
    pr_actions[optimal_action] += (1-epsilon)
    return pr_actions
```

### My main Q-Learning Function consists of 5 parts (inside each episode loop):

<br>

#### 1. Decision making based on policy which is updated based on the latest q-values

#### 2. Taking action based on decision just made, transition to next state, observing reward, saving it

#### 3. Updating state-action (q) values based on temporal reward just obtained and new state optimal action-value

#### 4. Updating States ( state <- next state ) 

#### 5. Ending episode (Termination) based on "done" flag received from "next_state_gen"

```python
def q_learning(env, revisit_punish = -20, alpha = 1, discount_factor = 0.9,
               alpha_decay = True, decay_mode = "exponential", decay_rate = 0.99,
               policy = "Softmax", beta = 1,increase_mode = False, increase_factor = 1.001,
               epsilon = 0.1, initial_state = 1, exploring_start = False):
    
    
    
    #Initialization: (All q-values to zero except terminal state):
    q_values = np.zeros((nS,nA))
    q_values[T] = np.ones(nA)*Rss[T]
    
    move_counts = []
    reward_sums = []
    for e in range(MIN_EPISODES):
        done = False #Termination Indicator
        
        if(e%5 ==1):
            alpha = alpha*0.99

        elif(decay_mode == "linear"):
            alpha = alpha - 1/MIN_EPISODES
            
        optimal_greedy_policy  = np.zeros((nS,nA))
        
        b = beta
        state = initial_state
        states = []
        move_count = 0 
        reward_sum = 0 
        count = 1 

        while(not(done)):  # Looping over each step in an epsiode:
            
            if((increase_mode == True)and(e % 10 == 1)):
                b = (b*increase_factor)
            ep = 0
            
            # 1. Making Decision based on current state-action(q) values:
            if(policy == "epsilon-greedy"):
                pr_actions = epsilon_greedy(q_values[state-1], epsilon)
                
            else:
                pr_actions = soft_max(q_values[state-1], b)
            
            decision = make_decision(pr_actions) 

            # 2. Taking action based on decision just made, Transition to Next State, Observing Reward, Saving Reward:
            next_state, done = next_state_gen(state, decision)
            
            if(next_state in states):
                ep = revisit_punish
            reward = Rss[next_state-1] + ep
            reward_sum += reward

            # 3. Updating state-action(q) values: (Predicting new q values for current state based on next_state)
            q_prediction = reward + discount_factor * max(q_values[next_state-1])
            error = q_prediction - q_values[state-1][decision]
            q_values[state-1][decision] += (alpha * error)

            # 4. Updating State:
            states.append(state)
            state = next_state
            move_count += 1

            # 5. Breaking in case of Termination, saving move_counts for future use
            if((done == True)or(count>1500)) :
                move_counts.append(move_count)
                reward_sums.append(reward_sum)
                for s in range(nS):
                    optimal_greedy_policy[s][np.argmax(q_values[s])] = 1
                break
        
            count +=1

    
    return move_counts, reward_sums, optimal_greedy_policy
```

```python
moves, total_reward, optimal_policy = q_learning(env)
```

```python
opt_action = np.where(optimal_policy ==1)[1]
opt_action = np.reshape(opt_action, (15,15))

print("Optimal Greedy Policy After 3000 Episodes(R2 = 0)")
for r in range(len(opt_action)):
    
    visualize = []
    for c in range(len(opt_action[0])):
        if(opt_action[r][c]==0):
            visualize.append("U")
            
        elif(opt_action[r][c]==1):
            visualize.append("R")
            
        elif(opt_action[r][c]==2):
            visualize.append("D")
            
        else:
            visualize.append("L")
    
    print(visualize)
```

<br>The <b> ϵ-greedy policy is a policy that chooses the best action (i.e. the action associated with the highest value) with probability 1−ϵ∈[0,1] and a random action with probability ϵ.</b> The problem with ϵ-greedy is that, <b> when it chooses the random actions (i.e. with probability ϵ), it chooses them uniformly </b>(i.e. it considers all actions equally good), even though certain actions (even excluding the currently best one) are better than others. Of course, this approach is not ideal in the case certain actions are extremely worse than others. Therefore,<b>  a natural solution to this problem is to select the random actions with probabilities proportional to their current values. These policies are called softmax policies.<br><br></b><br><b> Q-learning </b> is an <b> off-policy algorithm,</b> which means that, <b> while learning a so-called target policy,</b> it uses a so-called behaviour policy to select actions. <b> The behaviour policy can either be an ϵ-greedy, a softmax policy or any other policy that can sufficiently explore the environment while learning.</b>
