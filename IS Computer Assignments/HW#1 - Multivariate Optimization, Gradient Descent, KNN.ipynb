$$\Large \underline{{\textit{Cost Function Minimization using Gradient Descent:}}}$$

<p style = "font-family:cambria;">
    <font size="3">
        <br><b>Gradient Descent</b> has been developed as an optimiztion algorithm based on travelling from an initial position downhill towards a <b>local minimum point of a cost function</b> we want to minimize. Almost all Machine learning problems are eventually downsized to an optimiztion problem, which in we are willing to minimize a loss function, in order to make the best prediction possible.<br>The Iterative <b>Equation Describing a Gradient Descent Algorithm</b> is shown below: <br><br>$${\boxed{\displaystyle \mathbf {X} _{k+1}=\mathbf {X} _{k}+\alpha _{k}\,\mathbf {p} _{k}\,(X)}}$$<br><br>$\mathbf {X}: $ annotates the vector of variables forming the cost function  $\mathit f\,(\,{X}\,) \;;\; {X} = {{\begin{bmatrix}x_{1}\\\vdots \\x_{n}\end{bmatrix}}}$
<br>
 $\mathbf {X}^{k}: $ annotates the <b> value of variables after the k'th iteration</b> of the Gradient Descent Algorithm  $\mathit{X}^{k} = {{\begin{bmatrix}x_{1}^{k}\\\vdots \\x_{n}^{k}\end{bmatrix}}}$
<br><br><br>$\mathbf \alpha:$ shows the <b> learning rate of the algorithm </b> which plays the main role of <b> balancing the convergence rate with the risk of overshooting</b> the minimum on a Descent Algorithm.<br><br><font color= #CB4335><font><br><br> 

        
<p style = "font-family:georgia;">

<p style = "font-family:cambria;">
    <font size="3">
           * Well it's all <b>hidden in the descent direction $\mathit p_{k}\,(\,X\,)$:</b><br><br> $\mathbf p_{k}$ is a <b>descent(downhill)</b> direction which moving in that direction decreases the cost function and will ultimately walk us <b>towards a local minimum</b>.For a direction to satisfy the <b>descending condition</b>, it has to <b>form an angle more than $\mathit 90^{\circ}$ with the gradient direction</b> of the cost function at that particular position.<br><br>$$\boxed{\mathit {p_{k}}^{T}\;\nabla f_{k} \lt 0 }$$<br><br>So $\mathbf p_{k}$ can be decomposed to a matrix like $\mathit H_{k}$ and the Gradient Vector $\mathit \nabla f_{k}$ as follows:<br><br>$$\boxed{\mathit p_{k} = -\,H_{k}\,\nabla f_{k}}$$<br>* $\mathit H_{k}$ is a <b> positive-definite matrix </b> that corresponds to a <b>linear transformation which maps the gradient direction to the opposite ascent direction</b> of $\mathit p_{k}$ using the <b>transformation matrix $\mathit H_{k}$</b><br><br>Its obvious that if f, is a function of n variables $\begin{bmatrix}x_{1} & x_{2} & \cdots  & x_{n}\end{bmatrix}$, H should have n rows and n columns.<br><br>Here's where <b> Steepest Descent Algorithms </b> find their way out of their family "Descent Alogrithms", by putting $\mathbf H_{k} = I_{n\times n}:$ <br><br>$$\mathbf H_{k} = I_{n\times n}\Rightarrow\boxed{\mathit p_{k} = -\nabla f_{k}\,(\,X\,)}$$<br><br>Therefore the Equation shown Below can describe the <b> Dynamics of a Steepest Descent Algorithm:</b><br><br>$${\large\boxed{\displaystyle \mathbf {X} _{k+1}=\mathbf {X} _{k}-\alpha^{k}\; \nabla f(\mathbf {X} _{k})}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import path
```

$\large\textit Cost function  :$
$$\large\mathit \,f\,(\,x\,) = 3\,x_{1}\,^{2}\, + 2\,x_{1}\, + 4\,x_{2}\,^{2}\, + 8\,x_{2}\,$$

```python
def cost(x):
    return (3*x[0]**2 + 2*x[0] + 4*x[1]**2 + 8*x[1])
```

```python
x1 = np.linspace(-2.0, 1.0, 100)
x2 = np.linspace(-2.0, 1.0, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = (3*X1**2 + 2*X1 + 4*X2**2 + 8*X2)

levels = [-4.0,-2.0, 0.0, 2.0, 4.0, 8.0, 12.0,15.0]

plt.figure(figsize = (15,7))
cp = plt.contour(X1, X2, Y,levels)
plt.clabel(cp, inline=1, fontsize=10)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Contour plot of Cost Function f(x1,x2)')
plt.scatter([0,-1/3], [0,-1])
plt.annotate("  Initial Point", (0, 0))
plt.annotate("  Minimuim", (-1/3, -1))
plt.show()
```

<p style = "font-family:cambria;">

```python
def gradient(f,position,h = 0.0001):
    grad = np.zeros(len(position))
    next_position = np.zeros(len(position))

    for i in range(len(position)):
        next_poisiton = np.copy(position)
        next_position[:] = position[:]
        next_position[i] = next_position[i] + h
        grad[i] = (f(next_position) - f(position))/h
    return grad   
```

<br>         
<p style = "font-family:candara;">
    <font size="3">
        As there was an insist on implementing this assignment from <b> scratch </b> and not using <b> pre-constructed </b> functions and toolboxes, I assumed we have to implement the gradient operator too.The body of the function is shown in the cell above. It simply gets a function to take its derivative in every direction ( The Derivative is taken approximately ( $\mathit h = 10\,^{\,-4}$ ) )
        </p>

```python
def generate_nextpos(f,x,lr):
    y = np.zeros_like(x)
    grad = gradient(f,x)
    step = lr*gradient(f,x)
    y = x - step 
    return y,step,grad
```

<br>         
<p style = "font-family:candara;">
    <font size="3">
        The <b>"generate_nextpos"</b> function gets a <b> cost function, an input position and a learning rate,</b> calls the gradient function mentioned before, calculates the step size, and next position according to the given learning rate, and eventually <b>returns the gradient,the stepsize and next position on the cost function</b>
</p>

```python
def line_search(pos):
    lr_opt = (36*pos[0]**2 + 64*pos[1]**2 + 24*pos[0] + 128*pos[1] + 68)/(216*pos[0]**2 + 512*pos[1]**2 + 144*pos[0] + 1024*pos[1] + 536)
    return lr_opt
```

<br>         
<p style = "font-family:candara;">
    <font size="3">
        The <b> Line Search function</b> is implemented particularly for this cost function case and I have obtained the <b> optimal learing rate equation</b> based on <b> taking the derivative of the cost function with respect to $\alpha$ </b> as a <b> uni-variate optimiztion problem.</b> This equation is only drawn for this cost function and I implemented the function based of the final equation.I've written down the calculations in the "a"  part of this question.
</p>

```python
def armijo(pos,lr,c=0.0001,b=0.5):
    a1 = cost(generate_nextpos(cost,pos,lr)[0])
    a2 = c*lr*np.sum(gradient(cost,pos)**2)
    a3 = cost(pos)
    if(a1 > (a3 - a2)):
        print("updating lr\n")
        lr = (lr*b)
    return lr
```

<br>         
<p style = "font-family:candara;">
    <font size="3">
        The <b> armijo </b> function returns the <b> optimal learning rate for each step</b> based on the <b>armijo rule (wolphe conditions).</b> It gets the initial parameters $\mathit\,(\,c,\,\beta\,)\,$ and updates the learning rate if the wolphe conditions are satisfied. <b> The update is done by decresing the learning rate by a factor of $\beta$</b>
</p><br>

```python
MAX_ITERATION = 10000

def gradient_descent(f,pos,criteria_value = [0.001,1000], lr = 0.1, lr_mode = "constant", termination_criteria = "stepsize",details = False):
        
        position = np.zeros((len(pos),MAX_ITERATION))
        position[:,0] = np.array(pos)
        
        grad = np.zeros((len(pos),MAX_ITERATION))
        stepsize = np.zeros((len(pos),MAX_ITERATION))
        learning_rate = np.ones((MAX_ITERATION))*lr
        
        stepsize_criteria = criteria_value[0]
        iteration_criteria = int(np.asarray(criteria_value[1]))
        
        i=0
        while(True):
            
            
            if(lr_mode == "line_search"):
                learning_rate[i] = line_search(position[:,i])
                
            if(lr_mode == "armijo"):
                if(i==0):
                        learning_rate[i] = 1/(np.sqrt(np.sum(gradient(cost,pos)**2)))
                else:
                    learning_rate[i] = armijo(position[:,i],learning_rate[i-1],0.0001)
                
            position[:,i+1],stepsize[:,i],grad[:,i] = generate_nextpos(cost,position[:,i],learning_rate[i])
            
            if(details == True):
                print(f"Iteration No.{i}:\n\ngradient : {grad[:,i]}")
                print(f"position : {position[:,i]} \nloss = {cost(position[:,i])}\nstepsize : {stepsize[:,i]}\nnext: {position[:,i+1]}\n\n")    
            
            np.append(position,position[:,i+1])
            
            if(((termination_criteria == "stepsize") and ((stepsize[0,i] < stepsize_criteria) and (stepsize[1,i] < stepsize_criteria))) or ((termination_criteria == "iteration") and (i > iteration_criteria))):
                print("Terminated\n")
                res_position = position[:,i]
                res_stepsize = stepsize[:,i-1]
                res_iteration = i
                
                return(res_position,res_stepsize,res_iteration,position)
            
            i += 1
```

<br>         
<p style = "font-family:candara;">
    <font size="3">
    This is the <b> main function for implementing gradient descent</b> it calls the generate_nextpos function and initiates the learning rate in armijo rule mode, and of course decides when to terminate the algorithm based on the criteria value we have given to it. <b> Termination criteria mode</b>  can also be based on either <b> the total number of iterations </b> or <b> a lower bound for the step size we make in each step </b>

<br>$$\Large\mathbf {X} _{k+1}=\mathbf {X} _{k}-\alpha _{k}\nabla f(\mathbf {X} _{k});\ k\geq 0.\large\color{DarkBlue}{\;\;\;(\,I\,)}$$
    
<p style = "font-family:cambria;">
    <font size="3"><br>The eqaution shown above illustrates how a given point on a line,plane or hyperplane can move towards a local minumum point by moving in the opposite direction of the gradient vector

<p style = "font-family:cambria;">
    <b> Stochastic Descent Methods</b> are based on <b> optimizing a Finite Sum Approximation of the Cost function </b> instead of itself.The Equation Shown below demonstrates how a cost function such as $\mathit f\,(\,w\,)$ can be approximated by its average, $\mathit{\frac{1}{n}\sum _{i=1}^{n}f_{i}(w)}$ and instead of calculating the gradient of the cost function itself <b> we can calculate the gradient of its finite sum approximtion for each epoch</b>.
    </p>   
    
 $${\displaystyle f(w)={\frac {1}{n}}\sum _{i=1}^{n}f_{i}(w),}$$   

<p style = "font-family:cambria;">
    This<b>  Method of optimization </b> is helpful when we have <b> a lot of features </b>  that calculating the <b> analytical gradient</b> of the true loss function <b> requires a lot of time and computational resource</b> <br><br>For Example :<br><br> in an <b> SVM </b> problem with a high amount of features, the <b> hyper plane that acts as a classifier</b>  and is demonstrated by the weight vector $\mathbf W$ the <b> weights can be updated</b> in order to minimize the loss function without having it directly and <b> by only calculating loss gradients or subgradients ( for non-differentiable points ) with data collected from single data points across the dataset</b>. The equation below describes this method of weight and bias updating:<br><br>$${\displaystyle w:=w-\eta \nabla f\,(\,w\,)=w-{\frac {\eta }{n}}\sum _{i=1}^{n}\nabla f_{i}(w),}$$
    </p> 

```python
def lagrange_loss(x1,x2,landa,w):
    # Using "lambda" instead of "landa" for naming the parameter casuses primitive conflict.
    w = np.zeros_like(x1)
    loss = landa*(np.linalg.norm(w))**2 + np.average(max(0,1-x1*x2))
    return loss  
```

```python
def distance(x1, x2,dmode="Norm2"):
    dist = 0
    x1 = np.array(x1)
    x2 = np.array(x2)

    if(len(x1) != len(x2)):
        print("Dimension Mismatch!")
        return 0
    else:
        
        if(dmode == ("Norm1" or "Norm3")):
            dist = abs(x1-x2)

        else:
            dist = (x1-x2)**2

    return dist
```

      
<p style = "font-family:candara;">
    the fucntion <b> distance </b> calculates the <b> elementwise distance</b> , for cases we want to use <b> Euclidean Distance </b>, this function <b>returns a vector containing squares of elementwise difference of the 2 input vectors</b> and for cases <b> we want to use Manhattan Distance or ' p = $\inf$ distance ' this function returns elementwise difference.</b>
    </p>

```python
train_data = load_data("train_data.csv")
train_labels = load_data("train_labels.csv")
test_data = load_data("test_data.csv")
test_labels = load_data("test_labels.csv")
```

```python
def run_knn(train_data,train_labels,test_data,test_labels,k,dmode = "Norm2"):
    # distance deinition should be given to the function,otherwise it'll use Norm2 as default dist!
        
    # initializing necessary arrays:
    neighbours_dist = np.zeros((train_data.shape[0]))
    predicted_label = np.zeros(len(test_data))
    hit_array = np.zeros(len(test_data))
    #k_nearest_neighbours_dist = np.zeros((k,1))
    
    # searching for first k neighbours:
        
    for i in range(len(test_data)):
        
        for j in range(len(train_data)):
            
            if(dmode =="Norm1"):
                neighbours_dist[j] = np.sum(distance(test_data[i,:],train_data[j,:],dmode))
                
            elif(dmode == "Norm3"):
                neighbours_dist[j] = np.amax(distance(test_data[i,:],train_data[j,:],dmode))
                
            else:
                neighbours_dist[j] = np.sqrt(np.sum(distance(test_data[i,:],train_data[j,:],dmode)))
            
        k_nearest_neighbours = np.argsort(neighbours_dist)[::1][:k]
        #predicted_label[i] = np.bincount(k_nearest_neighbours).argmax()
        predicted_label[i] = np.bincount(train_labels[k_nearest_neighbours].flatten().astype(int)).argmax()
        #np.append(predicted_label,np.bincount(train_labels[k_nearest_neighbours].flatten().astype(int)).argmax())
        hit_array[i] = (test_labels[i] == predicted_label[i])
        
    acc = (np.sum(hit_array))/len(test_data)
            
    return acc,predicted_label
```

<p style = "font-family:candara;">
    This function <b> " run knn " </b> is the <b>main function</b> for this part. It <b>receives the train data and labels and the test data and their labels</b> and the <b> key parameter</b>  for a <b>knn algorithm</b> which is <b> k </b> and also a <b> distance mode</b> that will pass to its inner function 'distance' explained before.<br><br>
    <p style = "font-family:georgia;">
    <font size="5">
          <font color=#CB4335><font><b> Question </b>:<br><br><font color=#000000><font> <b>How is knn implemented ?</b><br>
    </font>    
</p><br>
        
<p style = "font-family:candara;"><font size="4">1. This function is <b> based on 2 nested loops</b>. The <b>Exterior</b> loop is performed on <b>test data</b>. The <b>Inner loop</b> is performed on <b>train data</b>. <br><br>2. Therefore when we the program is <b> fixed on for ex. the i'th test data</b> it will <b>calculate the distance of all the traning datapoints with itself in the inner loop and store them in an array called "neighbours_dist"</b>, which gathers the distance information of all the datapoints likely to be nighbours.<br><br> 3. Next the values in the array <b> ( the distances ) are sorted Ascending </b> and the <b> first k nearest neighbours indices are stored in another array called "k_nearest_neighbours"</b> <br><br>4. In the last step the function <b> finds the k corresponding lables to the k neighbours picked</b> and <b> assigns the mode label </b>( most frequent label ) to the predicted label for the test data ( predicted_label[i] )
    </font>
    </p>

The best classification results are obtained by the Manhattan Distance Metric ! $\Rightarrow$   <font color=#CB4335><font> <b> Justification and Interpretation:</b><font color=#000000><font> <br><br><br><br> <b> "Far from average" data can have a far more impact than close data on improving loss when a distance metric with $\mathit p \gt 1$ is used !</b> and this huge impact <b> can even grow Especially when the feature values haven't been normalized yet.</b> For instance;<br><br> Let's Imagine an un-normalized dataset a feature like height of adults will never have a difference bigger than 100 cm, but their salary can be different millions of dolars.When a feature like salary can cause a <b> huge differnece in the difference vector</b> of two humans ( acting as datas ) it can <b>scale the loss function by a factor of million!</b><br><br> This is when <b> Normalization becomes handy </b> and offers us <b> feature values</b> that are normally distibuted and <b>standardized across their mean</b>  and their distribution has a <b> similar level of Dispersion</b> .<br><br><br>  <font color=#CB4335><font> <b> So The important conclusion we can make from these results is :<br><br></b><font color=#000000><font> When <b> dealing with datasets</b> that <b>include very different types of features</b>, using <b>manhattan distance metric and normalizing features</b> can <b>improve the accuracy</b> by <b>decreasing the large effect of some out of range data</b> here and there in our dataset!<br>
