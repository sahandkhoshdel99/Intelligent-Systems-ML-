    <font>
</p>
        
<font color= #00000><font>
$${\boxed{\displaystyle \mathbf {X} _{k+1}=\mathbf {X} _{k}-\alpha^{k}\; \nabla f(\mathbf {X} _{k})}}$$
    
<br>
<p style = "font-family:cambria;">
    <font size="3">
        Our <b> Objective </b> is to <b> optimize the weights </b> in a <b> neural network </b> to <b> achieve the best prediction possible</b>  for our <b> regression problem </b>This can be done using <b> Back-Propagation </b>Back-Propagation lies on a mathematical property for differentiable functions called the <b> Chain Rule: </b> </b> 
            </p> 
<br><br>$${\displaystyle \boxed{{\frac {dC}{da^{L}}}\cdot {\frac {da^{L}}{dz^{L}}}\cdot {\frac {dz^{L}}{da^{L-1}}}\cdot {\frac {da^{L-1}}{dz^{L-1}}}\cdot {\frac {dz^{L-1}}{da^{L-2}}}\cdots {\frac {da^{1}}{dz^{1}}}\cdot {\frac {\partial z^{1}}{\partial x}}.}}$$<br>In the <b> Chain Rule </b> The derivatives (gradients) with respect to any weight in the network can be derived from the derivatives(gradients) with respect to next layer weights. The <b> Loss Function </b> is usually considered to be an <b> MSE </b> in regression problems.

```python
import numpy as np
def neuron(inputs,weights,bias):
    return (np.dot(inputs,weights)+bias)
```

<p style = "font-family:georgia;">
    <b><font size="3"> <font color="#6D81B9" >  How to initialize weights of a neural network? </b><font></font>
    </font>    
        </p>
    
<font color= #00000><font>
<p style = "font-family:cambria;">
    <font size="3">
        One solution is to use <b> "Xavier Initialization"</b>.This Weight initialization is based on the goal of making the <b> variance of the activations similar across every layer</b> . This constant variance helps <b> prevent the gradient from exploding or vanishing.</b><br>
<br>The <b> random weights </b> come from a <b> uniform distribution </b> in the following range:<br><br>$${W_{ij} \sim U\left[-\frac{1}{\sqrt{n}}, \frac{1}{\sqrt{n}}\right]}$$<br> Where <b> ${U}$ is the 'Uniform Distribution'</b>  and n is the <b>  "previous layer size'</b>


```python
def weight_initializer(prev_node_num,next_node_num):
    weights = np.zeros(prev_node_num*next_node_num)
    n = prev_node_num
    weights = np.random.uniform(-1/np.sqrt(n), 1/np.sqrt(n), size = prev_node_num*next_node_num)
    weights = weights.reshape(next_node_num,prev_node_num)
    return weights

weight_initializer(3,4)
```

```python
def tanh(x):
    return np.tanh(x)

def relu(x):
    out = np.zeros(len(x))
    for i in range(len(x)):
        out[i] = max(0,x[i])
    return out
```
